```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# cargar el archivo
url = '/content/MEDICAMENTOS_POS_20240413.csv'
df = pd.read_csv(url)

# mostrar las primeras filas
print(df.head())

# calcular estadistica descriptiva
print(df.describe())

# filtrar los datos
filtro = df['principioactivo'] == 'HIDROCLOROTIAZIDA'
datos_filtrados = df[filtro]
print(datos_filtrados)

# revisar cantidad de columnas y filas
shape = df.shape
print(shape)

# mostrar los ultimos datos del df
tail = df.tail()
print(tail)

# valores unicos
unique = df['nombrerol'].unique()
df['nombrerol'].drop_duplicates()
print(unique)

# contar valores
count_unique_values = df['nombrerol'].value_counts()
print(count_unique_values)


# resumen del df
print(df.info())

# ver elementos aleatorios
print(df.sample(5))

# construir un df del df
#df[['principioactivo', 'concentracion', 'factoresprecio']]

# crear una nueva columna
df['new'] = df['principioactivo'].apply(lambda x: len(str(x)))
print(df)
df.drop('new', axis = 'columns')
df.drop(1, axis= 'index')


```

```python
#segunda parte

plt.plot(df['viaadministracion'], df['cantidad'])
plt.xlabel ('viaadministracion')
plt.ylabel ('cantidad')
plt.title ('cantidad por viaadministracion')
plt.show ()
```

```python
#otras graficas

data = {'principioactivo': ['HIDROCLOROTIAZIDA', 'NEPAFENACO', 'DUTASTERIDA', 'GLICERINA', 'EZETIMIBE'],
        'cantidad': [12.5, 1, 0.5, 9, 10]}
df = pd.DataFrame(data)

#ordenar por ascendencia o descendencia
df_sorted = df.sort_values(by='cantidad', ascending=True)

#grafico de barras
plt.bar(df['principioactivo'], df['cantidad'])
plt.xlabel('principioactivo')
plt.ylabel('cantidad')
plt.title('cantidad por principioactivo')
plt.show()

#data = {'X': [1, 2, 3, 4, 5],
#        'Y': [2, 3, 5, 7, 9]}
#df = pd.DataFrame(data)

#grafico de dispersion
plt.scatter(df['principioactivo'], df['cantidad'])
plt.xlabel('principioactivo')
plt.ylabel('cantidad')
plt.title('Dispercion')
plt.show()

data = np.random.randn(1000)

#Histograma
plt.hist(data, bins=30)
plt.xlabel('principioactivo')
plt.ylabel('cantidad')
plt.title('Histograma')
plt.show()

#grafico de Torta
sizes = [15, 30, 45, 10]
values = ['principioactivo', 'cantidad', 'registrosanitario', 'viaadministracion']
plt.pie(sizes, labels=values, autopct='%1.1f%%')
plt.title('Grafico de Torta')
plt.show()


data = np.random.randn(100)
# grafico de area

data = {'principioactivo': [1, 2, 3, 4, 5],
        'cantidad': [2, 3, 5, 7, 9]}
df = pd.DataFrame(data)

plt.fill_between(df['principioactivo'],df['cantidad'])
plt.xlabel('principioactivo')
plt.ylabel('cantidad')
plt.title('Grafico de area')
plt.show()


# Grafico de barras apiladas
categories = ['principioactivo', 'cantidad', 'registrosanitario', 'viaadministracion']
values_one = [20, 45, 55, 77]
values_two = [50, 60, 70, 80]

plt.bar(categories, values_one, label = 'Grupo1')
plt.bar(categories, values_two, bottom=values_one, label = 'Grupo2')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Grafico de Barras apiladas')
plt.legend()
plt.show()
```
